# 2. Написать функцию, которая реализует численное решение задачи регрессии.


```python
def numerical_regression(x, y, learning_rate=0.01, iterations=1000):
    # Начальные значения весов и смещения
    m = 0
    b = 0
    n = len(x)

    # Итеративное обновление весов
    for _ in range(iterations):
        y_pred = m * x + b  # Предсказанные значения

        # Вычисление градиентов
        dm = (-2/n) * np.sum(x * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)

        # Обновление параметров модели
        m -= learning_rate * dm
        b -= learning_rate * db

    return m, b

# Примение  функции на данных
m, b = numerical_regression(x.flatten(), y)

# Предсказания
y_pred_numerical = m * x.flatten() + b
```
